### Using Binning to make a numeric column a categorical Value

```python
# We consider horsepower column
```

```python
print(df['horsepower'].max())
print(df['horsepower'].min())
```

```python
# Rough
np.linspace(0,100,5)
```

```python
# now we will make 3 bins : low ,medium, and high
```

```python
bins = np.linspace(df['horsepower'].min(), df['horsepower'].max(), 4)
```

```python
bins
```

```python
# naming the group
group_name = ['low','medium','high']
```

```python
df.columns
```

```python
df.shape
```

```python
df['horsepowere-bined'] = pd.cut(df['horsepower'],bins, labels = group_name, include_lowest = True)
```

```python
df['horsepowere-bined'].head()
```
